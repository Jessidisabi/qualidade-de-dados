```python
from IPython.core.display import publish_display_data
# bibliotecas de uso padrão
import pandas as pd
#biblioeca de analise
import numpy as np
#biblioteca de funções matematicas
import matplotlib.pyplot as plt
# biblioecas para graficos
```

```python
# os dados esão sujos
# carregar esses dados
df_dados_oscar = pd.read_csv('/content/exercicio_qualidade_dados_oscar.csv', on_bad_lines='warn')

#tenar ver dados
display(df_dados_oscar.head(20))
```

```python
# ver onde tem dado nulo e seus ipos
df_dados_oscar.info()
```

```python
# criar contagens
print(df_dados_oscar['Vencedor'].value_counts)
# dataframe utilizado [coluna].value_counts
```

```python
# remover tudo o que é duplicado
#guardando no sistema o valor atual de linhas
linhas_antes = len(df_dados_oscar)

#remove as duplicaas
#para remover vc chama a base de dados e fala q ela recebe ela mesma com o atributo drop_duplicaes()
df_dados_oscar = df_dados_oscar.drop_duplicates()
```

```python
display(df_dados_oscar.head(20))
```

```python
#mostrar valores unicos
print(df_dados_oscar['Vencedor'].unique())
```

```python
# vamos converter todos os dados para True ou False apenas
mapa_vencedores = {
    'Yes':True, 'Y': True, "No" : False, "True": True, '1': True,
    'Winner!': True, 'Sim' : True, 'TRUE': True, 'N' : False, '0': False,
    'nan': False
}

# trocando os dados
# chamamos a base de dados com a coluna, fazemos ela recebe ela receber ela msm
# com a coluna e usamos o atributo map para mapear os novos dados
df_dados_oscar['Vencedor'] = df_dados_oscar['Vencedor'].map(mapa_vencedores)
# para eliminar um dado nulo, vazio ou NaN
# chamamos a base de dados com a coluna, fazemos ela recebe ela receber ela msm
# com a coluna e usamo o atributo fillna('valor desejado)
df_dados_oscar['Vencedor'] = df_dados_oscar['Vencedor'].fillna(False)

display(df_dados_oscar.head(20))
```

```python
# corrigir dados escritos errados
# chamamos a base de dados, com a coluna, fazemos ela receber ela msm com a coluna
# e usamo o método replace('antes, depois')
df_dados_oscar['Indicado'] = df_dados_oscar['Indicado'].replace({'Oppenheimmer':'Oppenheimer'})
display(df_dados_oscar.head(2))
```
